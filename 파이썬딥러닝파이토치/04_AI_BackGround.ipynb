## 1. Universal Approximation Theorem
- 신경망 이론 중 가장 유명하고 신경망 학습의 특성을 잘 나타내주는 이론
- Hidden Layer 가 1개 이상이라는 것은 기본적인 MLP를 뜻한다.
- 다시 말해, 기본적인 MLP는 어떤 모델이든 만들 수 있다는 것을 의미한다.
- 하지만 학습 데이터 내에서 라는 전제가 있다 > 과적합의 개념과 이어진다.

## 2. 과적합
- 과적합은 신경망 모형뿐 아니라 AI 나 Data Science 측면에서 굉장히 큰 이슈이다.
- 이를 환벽하게 해결하는 방법이 없다.
- 다만 실험 설계를 통해 적절히 데이터를 나눠 과적합 정도를 파악하고 적절한 모델과 변수를 선택해야한다.

## 3. Gradient Vanishing Problem
- 기울기가 사라지는 현상이 하나의 문제점이다.
- 은닉층이 깊어질수록 가중치를 업데이트 할 때, 뒤에 작은 수를 곱하는 것을 반복하는데 0에 수렴해버리게 된다.
- 즉, Layer가 깊어질수록 Weight의 gradient는 큰 변화는 없어지고 Weight 의 변화도 거의 일어나지 않는다는 점

> 이를 해결하기 위한 다양한 활성화 함수가 나옴

## 4. 성능 지표
- 우리가 만든 성능을 측정할 필요가 있다.
- 일반적으로 loss값을 줄이기 위해 학습을 진행하는데, 사용자가 선택함에 따라, 다르게 설정할 수 있다.
### (1) MSE
- 기본 회귀 모형에서 많이 사용하는 Loss이자 성능 지표.
### (2) 정확도
- 분류문제에서 가장 많이 사용하는 성능지표
- (정상을 정상이라고 분류 + 오류를 오류라고 분류) / 전체 데이터의 수
### (3) 정밀도
- (정상을 정상이라고 분류) / (정상을 정상이라고 분류 + 오류를 정상이라고 분류)
### (4) 재현율
- (정상을 정상이라고 분류) / (실제로 정상인 데이터 수)
### (5) 특이도
- (불량을 불량이라고 분류) / 실제 불량인 데이터의 수
### (6) F1 Score
- Class Imbalance 상황에서 많이 사용하는 지표
- 2 * (정밀도 * 재현율) / (정밀도 + 재현율)
