```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Input, Dense, Flatten
from tensorflow.keras.optimizers import RMSprop
```

```python
# 시퀀스형
model = Sequential()
model.add(Flatten(input_shape = (28,28)))
model.add(Dense(units=5,activation = 'sigmoid'))
model.add(Dense(units=10, activation = 'softmax'))
model.summary()

model.compile(optimizer=RMSprop(learning_rate=0.1), loss = 'categorical_crossentropy')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
callback = EarlyStopping(monitor='val_loss', patience=5, verbose=2, mode = 'auto')
ret = model.fit(x_train, y_train, epochs=100, batch_size = 20, validation_split=0.2, verbose = 2, callbacks = [callback])
```

```python
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Input, Dense, Flatten
from tensorflow.keras.optimizers import RMSprop

# 시퀀스형
model = Sequential()
model.add(Flatten(input_shape = (28,28)))
model.add(Dense(units=5,activation = 'sigmoid'))
model.add(Dense(units=10, activation = 'softmax'))
model.summary()

model.compile(optimizer=RMSprop(learning_rate=0.1), loss = 'categorical_crossentropy')
```
